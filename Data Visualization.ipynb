```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(10, 100, size=len(categories))

x_data = np.linspace(0, 10, 100)
y_data = np.sin(x_data)

# Create the main figure
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line chart
axs[0, 0].plot(x_data, y_data, color='b')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Bar graph
axs[0, 1].bar(categories, values, color='g')
axs[0, 1].set_title('Bar Graph')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Pie chart
axs[1, 0].pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['orange', 'pink', 'lightblue', 'lightgreen'])
axs[1, 0].set_title('Pie Chart')

# Scatter plot
axs[1, 1].scatter(x_data, y_data, color='r')
axs[1, 1].set_title('Scatter Plot')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the dashboard
plt.show()
```
